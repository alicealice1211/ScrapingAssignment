## Structured Data from an API

For structured data, I chose to get a comprehensive exchange list from Coin Gecko(https://www.coingecko.com/) utilizing their API. 


It will be a straightforward example of scraping structured data. We will first make an API call then save all the data we get into a CSV file. 


Since I will be exporting the data into a CSV file, we need to first import CSV. As well as the “CoinGeckoAPI” package. 


```python
import csv
from pycoingecko import CoinGeckoAPI
```

After exploring Coin Geko’s API documentation (https://www.coingecko.com/en/api/documentation), I choose to scrape the complete exchange list which will include ID, name, year established, country, trade values, and many other details. 


Here I am making a quick connection with their API.

```python
cg = CoinGeckoAPI()
```

Here I am asking for their coin exchanges list.


```python
d = cg.get_exchanges_list()
```

I then created a short function to save the data into a CSV file with the name “struData.csv”. 


```python
def save():       
    file_name = "struData.csv"

    with open(file_name, 'w', encoding='utf-8', newline='') as csvfile:
        writ=csv.writer(csvfile)
        writ.writerow(d[0])
        for item in d:
            writ.writerow(item.values())
    print("csv file saved")   
```

Finally, I call the function to save the file. 


```python
save()
```
